```python
import numpy

sampling=100 #100개샘플확보
repeat=10 #10번반복
place_of_car=0 #차의 초기위치
place_of_car_std=1 #표준편차
speed_of_car=50 #차의속도
speed_of_car_std=20 #표준편차
particles=numpy.random.normal(place_of_car, place_of_car_std, sampling) #파티클 정규분포
for sampleitem in range(repeat):#10번 반복하
    noise=numpy.random.normal(0,speed_of_car_std, sampling) #노이즈발생기
    particles += speed_of_car + noise

print("sampleitem, particle positions: {}".format(sampleitem +1, particles[:10]))
```

```python
import numpy
import matplotlib.pyplot as plt

sampling=1000 #100개샘플확보
repeat=10 #10번반복
place_of_car=0 #차의 초기위치
place_of_car_std=1 #표준편차
speed_of_car=50 #차의속도
speed_of_car_std=20 #표준편차
particles=numpy.random.normal(place_of_car, place_of_car_std, sampling) #파티클 정규분포
plt.figure(figsize=(10, 6))
for sampleitem in range(repeat):#10번 반복하
    noise=numpy.random.normal(0,speed_of_car_std, sampling) #노이즈발생기
    particles += speed_of_car + noise

    plt.subplot(2, 5, sampleitem + 1)
    plt.hist(particles, bins=30, alpha=0.75)
    plt.title(f"Step {sampleitem + 1}")

plt.tight_layout()
plt.show()
```
